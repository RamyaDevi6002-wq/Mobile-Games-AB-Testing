```python
df.groupby("version")["sum_gamerounds"].mean()
```

```python
from scipy import stats
group_A=df[df["version"]=='gate_30']["sum_gamerounds"]
group_B=df[df["version"]=='gate_40']["sum_gamerounds"]
t_stat,p_value=stats.ttest_ind(group_A,group_B)
print("t_statistics:" ,t_stat)
print("P_value:", p_value)
```

```python
import pandas as pd
from scipy import stats

df = pd.read_csv("cookie_cats.csv")

retention_table = pd.crosstab(df["version"], df["retention_1"])
print("--- Retention Table ---")
print(retention_table)


chi2, p_val_chi, dof, ex = stats.chi2_contingency(retention_table)
print("\nChi-Square, P-Value:", p_val_chi)
```
